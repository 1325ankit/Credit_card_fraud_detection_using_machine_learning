```python
import numpy as np
from hpelm import ELM
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.metrics import log_loss
import pandas as pd



#X = df.drop('Class', axis=1)
featured_columns=['Time','Amount','V10','V11','V12','V14','V16','V17','V18']
X=df.loc[:,featured_columns]
y = df['Class']

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# Convert y_train to a NumPy array
y_train = np.array(y_resampled)

# Split the scaled data into training and test sets
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y_train, test_size=0.2, random_state=42)

# Create and train the ELM model
elm_model = ELM(X_train_scaled.shape[1], 1)
elm_model.add_neurons(800, 'sigm')  # Increase the number of neurons to 800
elm_model.train(X_train_scaled, y_train)  # No need for reshaping y_train

# Make predictions on the testing data
y_pred_elm = elm_model.predict(X_test_scaled)

# Generate threshold values
thresholds = np.arange(0.1, 1.0, 0.1)

# Convert ELM output to binary predictions for all thresholds
y_pred_binary = (y_pred_elm.reshape(-1, 1) > thresholds).astype(int)


# Calculate predicted probabilities for all thresholds
y_pred_proba = y_pred_elm.reshape(-1, 1)

# Calculate log loss for all thresholds
log_losses = [log_loss(y_test, y_pred_proba > threshold) for threshold in thresholds]

# Find the index of the threshold with the minimum log loss
best_threshold_idx = np.argmin(log_losses)

# Get the best threshold and corresponding binary predictions
best_threshold = thresholds[best_threshold_idx]
y_pred_best_threshold = y_pred_binary[:, best_threshold_idx]

# Calculate accuracy
elm_accuracy = accuracy_score(y_test, y_pred_best_threshold)

# Calculate precision, recall, and F1-score
elm_precision = precision_score(y_test, y_pred_best_threshold)
elm_recall = recall_score(y_test, y_pred_best_threshold)
elm_f1score = f1_score(y_test, y_pred_best_threshold)

# Print the metrics
print("ELM Accuracy:", elm_accuracy)
print("ELM Precision:", elm_precision)
print("ELM Recall:", elm_recall)
print("ELM F1-score:", elm_f1score)
```
